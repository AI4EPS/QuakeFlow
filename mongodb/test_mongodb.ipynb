```python
db = client["quakeflow"]
collection = db["waveform"]
```

```python
stream = obspy.read()
vec = []
for trace in stream:
    vec.append(trace.data)
vec = np.array(vec).T
print(vec.shape)
```

```python
post = {"_id":stream[0].id[:-1],
        "network":stream[0].stats.network,
        "station":stream[0].stats.station,
        "location":stream[0].stats.location,
        "channel":stream[0].stats.channel,
        "data": vec.tolist()}
```

```python
collection.insert_one(post)
```
